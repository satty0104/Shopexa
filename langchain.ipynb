```python
# Colab Cell 2
import os
import pandas as pd

# LangChain core
from langchain.schema import Document
from langchain.vectorstores import Chroma
from langchain.agents import Tool, initialize_agent
from langchain.agents.agent_types import AgentType

# Hugging Face integrations
from langchain_huggingface import HuggingFaceEmbeddings, HuggingFacePipeline
```

```python
# Colab Cell 3
# Upload 'products.csv' via Colab file browser or mount Drive.
#df = pd.read_csv("/content/drive/MyDrive/walmart/marketing_sample_for_walmart_com-product_details__20200101_20200331__30k_data.csv")
print(f"Loaded {len(df)} products.")
```

```python
df.columns
```

```python
# Colab Cell 4: Turn each row into a Document, embed on GPU
hf_embeddings = HuggingFaceEmbeddings(
    model_name="sentence-transformers/all-MiniLM-L6-v2",
    model_kwargs={"device": "cuda"},
)
docs = [
    Document(
        page_content=row["Description"],
        metadata={
            "url":       row["Product Url"],
            "name":      row["Product Name"],
            "price":     row["Sale Price"],
            "brand":     row["Brand"],
            "category":  row["Category"],
            "available": row["Available"],
        },
    )
    for _, row in df.iterrows()
]

# Build & persist the Chroma index
vectordb = Chroma.from_documents(
    documents=docs,
    embedding=hf_embeddings,
    persist_directory="chroma_db"
)
vectordb.persist()
```

```python
# Colab Cell 5: Create a Retriever
retriever = vectordb.as_retriever(search_kwargs={"k": 5})
```

```python
llm = HuggingFacePipeline.from_model_id(
    model_id="google/flan-t5-base",
    task="text2text-generation",
    # let Transformers auto‐map model to your GPU(s):
    model_kwargs={"device_map": "auto"},
    # tell the HF pipeline which CUDA device to use (0 = first GPU)
       pipeline_kwargs={"max_new_tokens": 256, "temperature": 0.7},
)
```

```python
# Colab Cell 5: Create a Retriever
retriever = vectordb.as_retriever(search_kwargs={"k": 5})
```

```python
# Colab Cell 7: Define a custom prompt for RetrievalQA
from langchain.prompts import PromptTemplate
from langchain.chains import RetrievalQA
prompt = PromptTemplate(
    input_variables=["context", "question"],
    template="""
You are a product-search assistant. Use the information from the retrieved products below
to answer the user's question as precisely as possible.

Product Info:
{context}

Question: {question}

Answer:
""".strip()
)

# Build the RetrievalQA chain
qa_chain = RetrievalQA.from_chain_type(
    llm=llm,
    chain_type="stuff",
    retriever=retriever,
    return_source_documents=True,
    chain_type_kwargs={"prompt": prompt}
)
```

```python
# Colab Cell 8: Run a test query
query = "Show me some cricket bats that are currently available"
output = qa_chain({"query": query})

print("Answer:\n", output["result"], "\n")
print("Source documents:")
for doc in output["source_documents"]:
    m = doc.metadata
    print(f"- {m['name']} | {m['brand']} | ₹{m['price']} | {m['category']} | Available: {m['available']}")
    print(f"  URL: {m['url']}\n  Snippet: {doc.page_content[:150].rstrip()}...\n")
```

```python
# prompt: show output from above code in a good manner using html after every query in a table format

from IPython.display import HTML, display
import pandas as pd

def display_output_as_html_table(output):
    """
    Displays the output of the qa_chain in a structured HTML table.
    Assumes the output is a dictionary with 'result' and 'source_documents'.
    """
    if not isinstance(output, dict) or 'result' not in output or 'source_documents' not in output:
        print("Warning: Output is not in the expected format for HTML display.")
        print(output)
        return

    html = "<h2>Agent Answer:</h2>"
    html += f"<p>{output['result']}</p>"

    html += "<h2>Source Documents:</h2>"
    html += "<table>"
    html += "<tr><th>Product Name</th><th>Brand</th><th>Price</th><th>Category</th><th>Available</th><th>URL</th><th>Snippet</th></tr>"

    for doc in output.get("source_documents", []):
        m = doc.metadata
        html += "<tr>"
        html += f"<td>{m.get('name', 'N/A')}</td>"
        html += f"<td>{m.get('brand', 'N/A')}</td>"
        html += f"<td>₹{m.get('price', 'N/A')}</td>"
        html += f"<td>{m.get('category', 'N/A')}</td>"
        html += f"<td>{m.get('available', 'N/A')}</td>"
        html += f'<td><a href="{m.get("url", "#")}" target="_blank">Link</a></td>'
        html += f"<td>{doc.page_content[:250].rstrip()}...</td>" # Increased snippet length
        html += "</tr>"

    html += "</table>"
    display(HTML(html))

# Example usage:
# Assuming 'qa_chain' is already defined from the preceding code
query = "Show me some headphones under $50"
output = qa_chain({"query": query})

# Display the output using the function
display_output_as_html_table(output)

# You can wrap this function call around any qa_chain execution
# Example with another query:
# query2 = "Tell me about products by the brand Apple that are available"
# output2 = qa_chain({"query": query2})
# display_output_as_html_table(output2)
```

```python

#execute

import pandas as pd
def get_season(date):
    """
    Determines the season based on the month.
    Args:
        date (pd.Timestamp): The date.
    Returns:
        str: The season ('Spring', 'Summer', 'Autumn', 'Winter').
    """
    month = date.month
    if 3 <= month <= 5:
        return 'Spring'
    elif 6 <= month <= 8:
        return 'Summer'
    elif 9 <= month <= 11:
        return 'Autumn'
    else:
        return 'Winter'

# Apply the function to create the 'Season' column
df['Season'] = df['Date and Time'].apply(get_season)

# Display the first few rows with the new 'Season' column
print(df.head())

# You can also check the distribution of seasons
print("\nSeason distribution:")
print(df['Season'].value_counts())
```
